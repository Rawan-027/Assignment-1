# Assignment 1: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset
df = pd.read_csv("Bakery Sales.csv")
# Display first 5 rows.
print("Dataset loaded successfully.")
print("First 5 rows:")
print(df.head())
```

```python
print("Data before cleaning:")
print(df.head())
```

### Check Missing Values

```python
# Check all values
print("Missing values before cleaning:")
print(df.isna().sum())
```

- Missing values were checked to ensure data quality.
- If missing values exist, they should be handled before analysis.

```python
df = df.fillna(0)

print("Missing values after filling:")
print(df.isna().sum())
```

Data Cleaning - Missing Values

- The dataset contained many missing values in product columns.
- These missing values do not represent missing data, but indicate that the product was not purchased.
- Therefore, missing values were replaced with 0.
- This allows accurate analysis of product sales.

### Check duplicate rows

```python
# checking duplicate rows
print("Shape before removing duplicates:", df.shape)
```

- Duplicate rows were checked to avoid repeated observations.
- Removing duplicates improves the accuracy of the analysis.

```python


df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

Data Cleaning - Duplicates

- The dataset contained duplicate rows (around 200).
- These duplicates were removed to improve data accuracy.
- Removing duplicates ensures that each transaction is counted only once.

### No. of rows and columns

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

- The dataset contains a suitable number of rows and columns for EDA.
- It includes enough records to identify patterns and trends.

### Data type of columns

```python
# viewing the data types of columns
df.dtypes
```

```python
# Statistical summary
df.describe(include='all')

print("Dataset information:")
print(df.info())
```

```python
#converting columns
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
df['Hour'] = df['datetime'].dt.hour
df['Month'] = df['datetime'].dt.month
df['Date'] = df['datetime'].dt.date

product_columns = [
    'angbutter', 'plain bread', 'jam', 'americano', 'croissant',
    'caffe latte', 'tiramisu croissant', 'cacao deep',
    'pain au chocolat', 'almond croissant', 'croque monsieur',
    'mad garlic', 'milk tea', 'gateau chocolat', 'pandoro',
    'cheese cake', 'lemon ade', 'orange pound', 'wiener',
    'vanila latte', 'berry ade', 'tiramisu', 'merinque cookies'
]
```

```python
print("Data after cleaning:")
print(df.head())
```

- The dataset includes both categorical and numerical information.
- This makes it suitable for different types of visual analysis.

### Descriptive summary Statistics

### Univariate Analysis

```python
sales_by_day = df.groupby('day of week')['total'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sales_by_day.plot(kind='bar')
plt.title("Total Sales by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

### Why this visualization?
- A bar chart helps compare total values across categories.

### Insight
- Certain days generate higher total revenue.
- This may indicate weekends or special days with higher demand.
- The bakery can plan promotions based on high-performing days.

```python
hourly_orders = df.groupby('Hour').size()

plt.figure(figsize=(10,6))
hourly_orders.plot(kind='bar')
plt.title("Number of Orders by Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Orders")
plt.show()
```

### Why this visualization?
- A bar chart is effective for showing activity across time intervals.

### Insight
- Some hours have significantly more orders than others.
- This highlights peak business hours.
- The bakery can adjust staffing during busy hours.

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='day of week', order=df['day of week'].value_counts().index)
plt.title("Number of Orders by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### Why this visualization?
- A countplot is useful for comparing frequency across categories.

### Insight
- Some days have more orders than others.
- This indicates variation in customer demand throughout the week.
- The bakery may experience peak days that require more resources.

```python
top_products = df[product_columns].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_products.plot(kind='bar')
plt.title("Top 10 Selling Products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

### Why this visualization?
- A bar chart is useful for comparing product performance.

### Insight
- A few products dominate sales.
- These items are the most popular among customers.
- The bakery can focus on these products to increase revenue.

```python
heatmap_data = df.pivot_table(
    index='day of week',
    columns='Hour',
    values='total',
    aggfunc='sum',
    fill_value=0
)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, annot=False, cmap='coolwarm')
plt.title("Heatmap of Sales by Day and Hour")
plt.xlabel("Hour")
plt.ylabel("Day of Week")
plt.show()
```

### Why this visualization?
- A heatmap helps identify patterns across two variables.

### Insight
- The heatmap shows peak sales times across days and hours.
- Certain combinations of day and hour have higher activity.
- This helps in planning staffing and inventory.

```python
top_dates = df.groupby('Date')['total'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_dates.plot(kind='bar')
plt.title("Top 10 Sales Dates")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

### Why this visualization?
- A bar chart highlights the highest-performing days.

### Insight
- Certain dates have significantly higher sales.
- These may correspond to events, weekends, or promotions.
- The bakery can analyze these days for better planning.

```python
plt.figure(figsize=(12,4))
sns.boxplot(data=df, x='total')
plt.title("Boxplot of Total Sales")
plt.xlabel("Total Sales")
plt.show()
```

### Why this visualization?
- A boxplot is useful for identifying the spread of sales values and detecting outliers.

### Insight
- The boxplot shows that most sales fall within a limited range.
- A few extreme values appear as outliers, indicating unusually large transactions.
- This suggests that the sales data is not evenly distributed.

```python
monthly_orders = df.groupby('Month').size()

plt.figure(figsize=(10,6))
monthly_orders.plot(kind='bar')
plt.title("Number of Orders by Month")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.show()
```

### Why this visualization?
- A bar chart shows trends across months.

### Insight
- Sales vary across months.
- Some months have higher customer activity.
- This may indicate seasonal trends.
